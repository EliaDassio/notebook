```python
f = open("./sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt","r")
text = " ".join(f.readlines())
special_char = ["\n", ".", ",", ";", ":", "!", "?", "(", ")"]
for c in special_char:
    text = text.replace(c, " ")
print(text)
f.close()
```

```python
frequence = {}
word_list = text.split(" ")
word_list = [word for word in word_list if word != ""]
print(word_list)
```

```python
for word in word_list:
    if word in frequence:
        frequence[word.lower()] = frequence[word.lower()] + 1
    else:
        frequence[word.lower()] = 1
print(frequence)
```

```python
for w in word_list:
    next_word[w.lower] = []
for index, word in enumerate(word_list):
    if index < len(word_list)-1:
        if word_list[(index + 1)] not in next_word[word.lower]:
            next_word[word.lower].append(word_list[index + 1])
print(next_word)
```
